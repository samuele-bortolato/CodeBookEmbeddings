```python
import torch
import pandas as pd
import numpy as np
```

```python
results1=torch.load('results\gloe_no_retrain\glue_results_no_retrain_seed42.pth')
results2=torch.load('results\gloe_no_retrain\glue_results_no_retrain_seed101.pth')
results3=torch.load('results\gloe_no_retrain\glue_results_no_retrain_seed255.pth')
results=[results1,results2,results3]
```

```python
text_result1={}
for k_runs, runs in results1.items():
    text_result1[k_runs]={}
    for k_run, run in runs.items():
        text_result1[k_runs][k_run]=f"{run[0]*100:,.1f}" if len(run)==1 else f"{run[0]*100:,.1f}/{run[1]*100:,.1f}"
df=pd.DataFrame(text_result1)
df
```

```python
text_result2={}
for k_runs, runs in results2.items():
    text_result2[k_runs]={}
    for k_run, run in runs.items():
        text_result2[k_runs][k_run]=f"{run[0]*100:,.1f}" if len(run)==1 else f"{run[0]*100:,.1f}/{run[1]*100:,.1f}"
df=pd.DataFrame(text_result2)
df
```

```python
text_result3={}
for k_runs, runs in results3.items():
    text_result3[k_runs]={}
    for k_run, run in runs.items():
        text_result3[k_runs][k_run]=f"{run[0]*100:,.1f}" if len(run)==1 else f"{run[0]*100:,.1f}/{run[1]*100:,.1f}"
df=pd.DataFrame(text_result3)
df
```

```python
tot={}
for task in results1.keys():
    if task not in tot:
        tot[task]={}
    for result in results:
        for run in result[task].keys():
            if run not in tot[task]:
                tot[task][run]={'mean':0, 'mean_squares':0}
            tot[task][run]['mean']+=np.array(result[task][run])*100
            tot[task][run]['mean_squares']+=np.square(np.array(result[task][run])*100)

text_tot={}
for task in tot.keys():
    text_tot[task]={}
    for run in tot[task].keys():
        tot[task][run]['mean']=tot[task][run]['mean']/len(results)
        tot[task][run]['mean_squares']=tot[task][run]['mean_squares']/len(results)
        tot[task][run]['variance']=tot[task][run]['mean_squares']-np.square(tot[task][run]['mean'])
        mean=tot[task][run]['mean']
        std=np.sqrt(tot[task][run]['variance'])
        text_tot[task][run]=f"{mean[0]:,.1f}\u00B1{std[0]+0.05:,.1f}" if len(mean)==1 else f"{mean[0]:,.1f}\u00B1{std[0]+0.05:,.1f} / {mean[1]:,.1f}\u00B1{std[1]+0.05:,.1f}"

df=pd.DataFrame(text_tot)
df
```
